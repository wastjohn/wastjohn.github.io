```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ripser import Rips

data1 = pd.read_csv('data/points1.csv', header=None)
data2 = pd.read_csv('data/points2.csv', header=None)
data3 = pd.read_csv('data/points3.csv', header=None)
data4 = pd.read_csv('data/points4.csv', header=None)
data5 = pd.read_csv('data/points5.csv', header=None)
data6 = pd.read_csv('data/points6.csv', header=None)
data11 = pd.read_csv('data/points11.csv', header=None)
data12 = pd.read_csv('data/points12.csv', header=None)
data13 = pd.read_csv('data/points13.csv', header=None)
data14 = pd.read_csv('data/points14.csv', header=None)
data21 = pd.read_csv('data/points21.csv', header=None)
data22 = pd.read_csv('data/points22.csv', header=None)
data4001 = pd.read_csv('data/points400_1.csv', header=None)
data4002 = pd.read_csv('data/points400_2.csv', header=None)

data_list = [data1, data2, data3, data4, data5, data6,
             data11, data12, data13, data14, 
             data21, data22,
             data4001, data4002]
```

```python
for i in data_list[:6]:
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,2)
    plt.scatter(i[0], i[1])
    plt.subplot(1,2,1)
    rips = Rips(maxdim=2)
    dgm_noisy = rips.transform(i)
    plt.title('Persistence Diagram')
    rips.plot()

for i in data_list[6:10]:
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,2, projection='3d')
    plt.scatter(i[0], i[1], i[2])
    plt.subplot(1,2,1)
    rips = Rips(maxdim=2)
    dgm_noisy = rips.transform(i)
    plt.title('Persistence Diagram')
    rips.plot()

for i in data_list[10:]:
    rips = Rips(maxdim=2)
    dgm_noisy = rips.transform(i)
    plt.title('Persistence Diagram')
    rips.plot(show=True)
```

```python
rips1 = Rips(maxdim=2)
rips1.transform(data1)

rips2 = Rips(maxdim=2)
rips2.transform(data2)

rips3 = Rips(maxdim=2)
rips3.transform(data3)

rips4 = Rips(maxdim=2)
rips4.transform(data4)

rips5 = Rips(maxdim=2)
rips5.transform(data5)

rips6 = Rips(maxdim=2)
rips6.transform(data6)

rips11 = Rips(maxdim=2)
rips11.transform(data11)

rips12 = Rips(maxdim=2)
rips12.transform(data12)

rips13 = Rips(maxdim=2)
rips13.transform(data13)

rips14 = Rips(maxdim=2)
rips14.transform(data14)

rips21 = Rips(maxdim=2)
rips21.transform(data21)

rips22 = Rips(maxdim=2)
rips22.transform(data22)

rips4001 = Rips(maxdim=2)
rips4001.transform(data4001)

rips4002 = Rips(maxdim=2)
rips4002.transform(data4002)
```
